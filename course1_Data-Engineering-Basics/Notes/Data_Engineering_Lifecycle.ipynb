# Data Engineering Lifecycle

## Data Platform Architecture Layers

A **layer** represents functional components that perform specific tasks within a data platform. The key layers include:

## 1. Data Ingestion (Data Collection) Layer
Responsible for connecting to source systems and bringing data into the platform.

### Key Tasks:
- Connect to data sources.
- Transfer data in streaming, batch, or both modes.
- Maintain metadata about collected data.

### Tools:
- **Batch & Streaming:** Google Cloud DataFlow, IBM Streams, IBM Streaming Analytics, Amazon Kinesis, Apache Kafka.

## 2. Data Storage and Integration Layer
Stores and integrates ingested data for processing and long-term use.

### Key Responsibilities:
- Store data for processing and long-term retention.
- Transform and merge extracted data.
- Support both streaming and batch data availability.
- Ensure reliability, scalability, performance, and cost-efficiency.

### Storage Technologies:
- **Relational Databases:** IBM DB2, Microsoft SQL Server, MySQL, Oracle, PostgreSQL.
- **Cloud Relational Databases (DBaaS):** IBM DB2 on Cloud, Amazon RDS, Google Cloud SQL, SQL Azure.
- **NoSQL Databases:** IBM Cloudant, Redis, MongoDB, Cassandra, Neo4J.

### Integration Tools:
- **Enterprise Integration:** IBM Cloud Pak for Data, IBM Cloud Pak for Integration, Talend Data Fabric, Open Studio.
- **Open Source:** Dell Boomi, SnapLogic.
- **Cloud Integration (iPaaS):** Adeptia Integration Suite, Google Cloud’s Cooperation 534, IBM’s Application Integration Suite, Informatica Integration Cloud.

## 3. Data Processing Layer
Validates, transforms, and applies business logic to data.

### Responsibilities:
- Read data in batch or streaming modes.
- Apply transformations such as structuring, normalization, denormalization, and data cleaning.
- Support querying tools and programming languages.
- Scale for growing datasets.

### Processing Technologies:
- **Data Transformation Tools:** Spreadsheets, OpenRefine, Google DataPrep, Watson Studio Refinery, Trifacta Wrangler.
- **Programming Libraries:** Python (pandas, NumPy), R.
- **Big Data Processing:** Hadoop File Distribution System (HDFS), Apache Spark.

### Processing & Storage Overlap:
- In **relational databases**, storage and processing occur in the same layer.
- In **Big Data systems**, data is first stored (e.g., HDFS) before being processed.

## 4. Analysis and User Interface Layer
Delivers processed data to consumers.

### Data Consumers:
- **Business Analysts & Stakeholders** – Use dashboards and reports.
- **Data Scientists & Analysts** – Perform further data processing.
- **Other Applications & Services** – Consume data for various uses.

### Key Capabilities:
- **Querying Tools:** SQL (relational databases), SQL-like tools (e.g., CQL for Cassandra).
- **Programming Languages:** Python, R, Java.
- **APIs:** For reports, online/offline processing, real-time data consumption.
- **Dashboarding & BI Tools:** IBM Cognos Analytics, Tableau, Jupyter Notebooks, Microsoft Power BI.

## 5. Data Pipeline Layer (Overlaying Layers)
Implements and maintains a continuously flowing data pipeline using **Extract, Transform, Load (ETL)** tools.

### Pipeline Solutions:
- **Apache Airflow**
- **Google Cloud DataFlow**

## Data Store Design Considerations

A **data store**, or **data repository**, refers to data that has been **collected, organized, and isolated** for business operations, reporting, and analysis. It can take various forms, such as:

- **Databases**
- **Data Warehouses**
- **Data Marts**
- **Big Data Stores**
- **Data Lakes**

A well-designed data repository ensures **scalability** and **performance** under high workloads.

### 1. Key Design Considerations

#### 1.1 Type of Data
The type of data determines the best storage solution:

- **Structured data** → Best stored in **Relational Databases (RDBMS)**
- **Semi-structured & unstructured data** → Best suited for **Non-Relational Databases (NoSQL)**

##### 1.1.1 NoSQL Database Types:
- **Key-Value Stores** (e.g., Redis)
- **Document Stores** (e.g., MongoDB)
- **Column Stores** (e.g., Cassandra)
- **Graph Databases** (e.g., Neo4J)

Each type is optimized for different **querying** and **processing** needs.

#### 1.2 Data Volume & Scalability
- **Data Lakes** store large volumes of raw data in its native format.
- **Big Data Repositories** handle **high-volume, high-velocity, diverse data** using **distributed processing** (e.g., Hadoop).
- **Scalability** ensures the system can handle growing **data, workloads, and users**.

#### 1.3 Intended Use
The purpose of data storage impacts database design:

##### 1.3.1 Transactional Systems (OLTP)
- Require **high-speed read, write, and update** operations.
- **Normalization** optimizes efficiency and storage usage.

##### 1.3.2 Analytical Systems (OLAP)
- Execute **complex queries on large datasets** for reporting & insights.
- **Denormalization** improves query performance.

#### 1.4 Storage Considerations
Key factors affecting storage design:

- **Performance** → Throughput (read/write speed) & latency (response time).
- **Availability** → Ensures **zero downtime** for data access.
- **Integrity** → Protects against **data corruption, loss, and cyber threats**.
- **Recoverability** → Ensures **data restoration** after failures or disasters.

#### 1.5 Privacy, Security & Governance
- **Access control, encryption, and monitoring** ensure secure data management.
- Regulations like **GDPR, CCPA, HIPAA** enforce data ownership, use, and protection.
- **Data governance strategies** must be **integrated from the start** to avoid security patchwork.

## Enterprise-Level Data Platform Security

Enterprise-level data platforms and repositories must address security at multiple levels. Security in data platforms encompasses **physical infrastructure security, network security, application security, and data security**.

### 1. The CIA Triad: Foundation of Security
An effective **information security strategy** follows the **CIA triad**:

- **Confidentiality (C)** – Prevents unauthorized access to data.
- **Integrity (I)** – Ensures data is trustworthy and unaltered.
- **Availability (A)** – Guarantees authorized users can access data when needed.

This model applies across **infrastructure, network, application, and data security**.

## 2. Security Layers in Data Platforms

### **Physical Infrastructure Security**
Physical security is crucial for IT systems and cloud computing infrastructure. Key measures include:

- **Facility access control** with authentication and 24/7 surveillance.
- **Power redundancy** via multiple power sources, generators, and battery backups.
- **Climate control** for temperature and humidity regulation.
- **Disaster prevention** – Facilities are placed in safe locations (e.g., avoiding flood-prone areas).
- **Structural resilience** – Earthquake-resistant structures, multi-level lightning protection, and grounding systems.

### **Network Security**
Network security protects interconnected systems and data. Key solutions include:

- **Firewalls** – Block unauthorized internet access.
- **Network access control (NAC)** – Ensures only authorized devices connect.
- **Network segmentation** – Creates virtual LANs to isolate critical assets.
- **Security protocols** – Encrypts data in transit to prevent interception.
- **Intrusion Detection & Prevention Systems (IDPS)** – Monitors and mitigates network threats.

### **Application Security**
Application security safeguards customer data and ensures optimal performance. Key practices include:

- **Threat modeling** – Identifies vulnerabilities and attack patterns.
- **Secure design** – Mitigates risks at the foundational level.
- **Secure coding** – Follows best practices to prevent vulnerabilities.
- **Security testing** – Identifies and fixes issues before deployment.

### **Data Security**
Data security ensures protection for both **data at rest** and **data in transit**:

- **Authentication & Authorization** – Restricts data access based on user roles.
- **Data Encryption** – Protects stored data (e.g., in databases, backups) and transmitted data (e.g., via HTTPS, SSL, TLS).

## 3. Security Monitoring & Compliance
Enterprises must **proactively track, monitor, and react** to security incidents. Key strategies include:

- **End-to-end security visibility** – Integrating security tools across the enterprise.
- **Audit trails & compliance** – Maintaining detailed logs for security investigations.
- **Real-time security alerts** – Enabling timely responses to threats.

---

### Conclusion
Enterprises must implement a **corporate-level security policy** that aligns **business, IT, and security stakeholders** through **people, policies, processes, systems, and tools**. Ensuring **end-to-end data protection** is essential for maintaining trust, compliance, and operational integrity.


## Importance of Data Security

## The Critical Importance of Data Security

If **data is not secure, nothing else matters** in the long run. Data breaches can impact companies of all sizes, and for smaller organizations, **losing access to data can be a company-ending problem**. Security must be a **priority at every stage**, not an afterthought before deployment.

The *Economist* once stated that **data has surpassed oil as the world's most valuable resource**. Given its immense value, **securing and protecting data** is a fundamental requirement for organizations. A failure in data security can lead to **catastrophic consequences**.

Data security is **not just the responsibility of data engineers**—it must be embedded into **data architecture, organizational processes, and company-wide policies**. Organizations must ensure:

- **Production and non-production environments remain separate** to prevent sensitive data exposure.
- **Access to production data is strictly limited** to those who genuinely need it.
- **Users have the least privilege necessary** to perform their roles.

Excessive access increases the risk of **data breaches and security vulnerabilities**.

While external cyberattacks are a concern, **most data threats come from within the organization**. High-profile cases include **unauthorized access to sensitive data**, such as employees in the healthcare sector viewing **celebrity prescription records**.

### **Data as a Natural Resource**
In the era of **big data**, information is **as valuable as oil**. Organizations must ensure **financial protection and risk mitigation** by implementing **robust security protocols**. A **data breach** can not only **damage reputation** but potentially **shut down an organization**.

An often-overlooked aspect of security is **data restoration and disaster recovery**. Hardware failures should not result in **permanent data loss**. A **secure data strategy** includes:

- **Reliable backups** to restore lost information.
- **Disaster recovery plans** to mitigate data loss risks.

## Data Wrangling

## Data Wrangling and Transformation

### Data Wrangling Overview
Raw data undergoes transformations and cleansing to be analytics-ready. This process, called data wrangling or data munging, involves:

- **Exploration** – Understanding data structure and contents  
- **Transformation** – Changing format, schema, or structure  
- **Validation** – Checking for inconsistencies and errors  
- **Preparation** – Making data suitable for analysis  

---

## Key Data Transformations

### Structuring Data
Transformations that change the form and schema of data include:

#### **Joins vs. Unions**
- **Joins**: Combine columns from multiple tables into a single row  
- **Unions**: Combine rows from multiple tables into a single dataset  

#### **Normalization vs. Denormalization**
- **Normalization**: Reduces redundancy and improves consistency (common in transactional systems)  
- **Denormalization**: Combines data from multiple tables for faster querying (common in analytical systems)  

---

## Data Cleaning

### Common Data Issues and Fixes

#### 1. **Missing Values**
- Remove records with missing values  
- Source missing data from other sources  
- **Imputation** – Calculate missing values using statistical methods  

#### 2. **Duplicate Data**
- Detect and remove duplicate records  

#### 3. **Irrelevant Data**
- Remove data not relevant to the analysis (e.g., contact numbers in a health study)  

#### 4. **Data Type Conversion**
- Ensure values are stored in the correct data type (e.g., convert text to numerical or date format)  

#### 5. **Standardization**
- Convert text to lowercase  
- Standardize date formats and units of measurement  

#### 6. **Syntax Errors**
- Remove extra spaces, typos, and inconsistent formatting (e.g., "New York" vs. "NY")  

#### 7. **Outliers**
- Detect errors (e.g., a voter’s age listed as 5)  
- Analyze outliers to determine whether they should be removed or kept  

---

### Data Profiling
- Examines structure, relationships, and inconsistencies in data  
- Detects anomalies like missing values, duplicates, or incorrect ranges  

### Data Visualization
- Graphs and charts help identify patterns and outliers  
- **Example**: Plotting income distribution to spot unusual values

## Data Analysis Techniques

These techniques include counting and aggregating your data set, identifying extreme values, slicing data, sorting data, filtering patterns, and grouping data. In this context, SQL querying techniques are often applied, but the methods also work for NoSQL databases (like Cassandra CQL and Cypher for Neo4J) and APIs for data querying.

### Counting Data

- **COUNT()**: Used to count the number of rows (records) in a data set.  
- **DISTINCT()**: Filters unique values.  
- **COUNT() + DISTINCT()**: Combines both to count unique records (e.g., unique car dealers).

### Aggregation Functions

- **SUM()**: Sums the values in a numeric column.
- **AVG()**: Calculates the average value in a numeric column.
- **STDDEV()**: Measures the spread of values in a column, showing the variation of data.

Example: In a car data set, the average price of a used car is less than $6,000, but the standard deviation is over $11,000, indicating a wide spread in prices.

### Identifying Extreme Values

- **MAX()**: Retrieves the maximum value in a column (e.g., highest purchase amount).
- **MIN()**: Retrieves the minimum value in a column (e.g., lowest purchase amount).

### Slicing Data

- **SLICE()**: Allows you to retrieve data based on specific conditions (e.g., customers living in a certain area or with a specific spending range).
- **Operators**: Use multiple conditions for slicing data, such as filtering customers who spent between $1,000 and $2,000 and live in a certain region.

### Sorting Data

- **ORDER BY()**: Sorts data in ascending or descending order, such as arranging sales by date to spot trends (e.g., sales during festival times).

### Filtering Patterns

- **LIKE()**: Performs partial matches using patterns.  
Example: You can filter customer data based on the first three digits of their pincode using the `LIKE` operator with wildcard characters.

### Grouping Data

- **GROUP BY**: Groups data by specific columns to summarize results.  
Example: Grouping customer spending by pincode and using `SUM()` to calculate total spending in each area.

These querying techniques are essential for analyzing data effectively and can be applied across different querying languages, including SQL and NoSQL databases.


## Performance Tuning and Troubleshooting for Data Pipelines and Databases

One of the key responsibilities of a data engineer is to monitor and optimize systems and data flows for performance and availability. In a data engineering lifecycle, performance impacts various areas.

### Data Pipeline Performance

Data pipelines represent the journey of data from source to destination, passing through multiple systems, applications, and processes. Performance issues in data pipelines can arise due to:

- **Scalability**: Handling increasing data sets and workloads.
- **Application Failures**: System crashes or errors.
- **Scheduled Jobs**: Delays, dependency issues, or tasks not running as expected.
- **Tool Incompatibilities**: Problems arising from using various tools for different tasks.

### Performance Metrics for Data Pipelines

To evaluate data pipeline performance, these metrics are essential:

- **Latency**: Time taken for a service to fulfill a request.
- **Failures**: The failure rate of services.
- **Resource Utilization**: Patterns of resource consumption.
- **Traffic**: Number of user requests within a given period.

### Troubleshooting Steps for Performance Issues

Steps for resolving performance issues in data pipelines:

1. **Collect Information**: Gather incident details and confirm if it's a real issue.
2. **Check Software Versions**: Ensure the right versions of software and source codes are in use.
3. **Review Logs and Metrics**: Identify infrastructure, data, or software issues.
4. **Test in a Staging Environment**: If logs aren't helpful, reproduce the issue in a test environment.
5. **Root Cause Analysis**: Validate the cause of the issue through testing and move to production accordingly.

### Database Performance Metrics

For databases, performance metrics include:

- **System Outages**: Monitoring downtimes.
- **Capacity Utilization**: Ensuring resources are not overburdened.
- **Application Slowdown**: Detecting performance issues.
- **Query Performance**: Measuring the efficiency of queries.
- **Conflicting Activities**: Managing multiple queries or requests from users and batch activities.

### Best Practices for Database Optimization

To optimize database performance, consider the following:

- **Capacity Planning**: Determine the resources required for performance, factoring in future growth.
- **Database Indexing**: Speeds up data retrieval by reducing the need for full-table scans.
- **Database Partitioning**: Divides large tables into smaller sections for faster queries and better manageability.
- **Database Normalization**: Minimizes data redundancy and improves querying efficiency.

### Monitoring and Alerting Systems

Monitoring tools provide real-time visibility into system performance:

- **Database Monitoring Tools**: Track performance indicators through snapshots to identify when issues started.
- **Application Performance Management Tools**: Measure request times, error messages, and resource usage to allocate resources more efficiently.
- **Query Performance Monitoring**: Track query throughput and performance for better resource planning.
- **Job-Level Runtime Monitoring**: Break down jobs into steps and monitor their completion times to avoid costly failures.

### Maintenance Routines

Regular maintenance ensures optimal system operation:

- **Preventive Maintenance**: Identifies faults and low availability by generating data from routine checks.
- **Time-Based**: Scheduled activities at pre-set intervals.
- **Condition-Based**: Performed when specific issues or performance drops are noticed.


## Data Governance

Data Governance is a collection of principles, practices, and processes to maintain the security, privacy, and integrity of data through its lifecycle. A data governance framework encompasses every part of an organization's data management process, including technologies, databases, and data models.

### Types of Data Protected
- Personal and sensitive data, such as information that can identify an individual or cause harm (e.g., race, sexual orientation, genetic data).

### Key Regulations
- **GDPR (General Data Protection Regulation)**: Protects personal data and privacy of EU citizens for transactions within EU member states.
- **CCPA (California Consumer Privacy Act)**: Protects customer data in California.
- Industry-specific regulations include:
  - **HIPAA** for healthcare data
  - **PCI DSS** for credit card data
  - **SOX** for financial information

### Compliance
Compliance refers to processes and procedures ensuring an organization adheres to regulations and conducts operations legally and ethically. Non-compliance can lead to penalties, damage to reputation, and loss of trust. It is an ongoing process requiring people, processes, and technology.

## Data Lifecycle Stages

1. **Data Acquisition**:
   - Define the data to collect, establish contracts, and obtain consent.
   - Set the intended use and limit the data collected to only what is necessary.
   
2. **Data Processing**:
   - Establish how data will be processed and the legal basis for this (e.g., consent or contract).

3. **Data Storage**:
   - Determine where data will be stored and implement security measures to prevent breaches.

4. **Data Sharing**:
   - Identify third-party vendors with data access and ensure they comply with regulations.

5. **Data Retention and Disposal**:
   - Define policies for retention and deletion of data, ensuring complete removal from all locations.

6. **Audit Trail**:
   - Maintain a verifiable trail for data acquisition, processing, storage, access, retention, and deletion.

## Tools and Technologies for Compliance

### Authentication and Access Control
- Layered authentication (passwords, tokens, biometrics) for data protection.
- Role-based access control to restrict data access based on user roles.

### Encryption and Data Masking
- **Encryption**: Data is encoded to be accessible only through secure decryption.
  - Available for both **data at rest** (in storage) and **data in transit** (moving through systems).
- **Data Masking**: Anonymizes or pseudonymizes data for downstream processing.
  - **Anonymization**: Abstracts data at the presentation layer.
  - **Pseudonymization**: Replaces personal identifiers with artificial identifiers.

### Hosting Compliance
- On-premise and cloud systems must comply with international data transfer restrictions.

### Monitoring and Alerting
- **Security Monitoring**: Tracks and reacts to security violations across infrastructure and applications.
- **Alerting**: Flags breaches in real-time, triggering remedial actions based on severity.

### Data Erasure
- Data erasure is a method of permanently deleting data by overwriting, making it irretrievable.

## Conclusion
Data governance and compliance are critical throughout the data lifecycle. Tools and technologies help ensure governance implementation, securing sensitive data, maintaining privacy, and adhering to legal requirements.


## Overview of the DataOps Methodology

Gartner defines DataOps as a collaborative data management practice focused on improving the communication, integration, and automation of data flows between data managers and consumers across an organization. DataOps aims to create predictable delivery and change management of data, data models, and related artifacts. DataOps uses technology to automate data delivery with the appropriate levels of security, quality, and metadata to improve the use and value of data in a dynamic environment.”

(Source: https://blogs.gartner.com/nick-heudecker/hyping-dataops/)

A small team working on a simpler or limited number of use cases can meet business requirements efficiently. As data pipelines and data infrastructures get more complex, and data teams and consumers grow in size, you need development processes and efficient collaboration between teams to govern the data and analytics lifecycle. From data ingestion and data processing to analytics and reporting, you need to reduce data defects, ensure shorter cycle times, and ensure 360-degree access to quality data for all stakeholders.

DataOps helps you achieve this through metadata management, workflow, and test automation, code repositories, collaboration tools, and orchestration to help manage complex tasks and workflows. Using the DataOps methodology ensures all activities occur in the right order the right security permissions. It helps set in a continual process that allows you to cut wastages, streamline steps, automate processes, increase throughput, and improve continually. Several DataOps Platforms are available in the market, some of the popular ones being IBM DataOps, Nexla, Switchboard, Streamsets, and Infoworks.

#### DataOps Methodology:
The purpose of the DataOps Methodology is to enable an organization to utilize a repeatable process to build and deploy analytics and data pipelines. Successful implementation of this methodology allows an organization to know, trust, and use data to drive value.

It ensures that the data used in problem-solving and decision-making is relevant, reliable, and traceable and improves the probability of achieving desired business outcomes. And it does so by tackling the challenges associated with inefficiencies in accessing, preparing, integrating, and making data available.

In a nutshell, the DataOps Methodology consists of three main phases:

 * The Establish DataOps Phase provides guidance on how to set up the organization for success in managing data.

 * The Iterate DataOps Phase delivers the data for one defined sprint.

 * The Improve DataOps Phase ensures learnings from each sprint is channeled back to continually improve the DataOps process.
